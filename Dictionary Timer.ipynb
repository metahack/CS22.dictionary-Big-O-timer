```python
# Earle Bowers CS22 F14 LBCC
# This program confirms that dictionary get and set are constant time operations

import timeit
import random

test_counts = []
test_set_data = []
test_get_data = []

def test_set(dict_to_use, i):
    dict_to_use[i] = True

def test_get(dict_to_use, i):
    return dict_to_use[i]

#EZ range = 100,000 to 5,000,000 in steps of 100,000
for i in range(100 * 1000, 5000 * 1000 + 1, 100 * 1000):

    dictionary = {}
    for x in range(0,i):
        dictionary[x] = False
        
    rand = random.randint(0,i)

    s = timeit.Timer('test_set(dictionary, %i)' % rand, 'from __main__ import test_set, dictionary')
    set_time = s.timeit()

    rand = random.randint(0,i)

    g = timeit.Timer('test_get(dictionary, %i)' % rand, 'from __main__ import test_get, dictionary')
    get_time = g.timeit()

    test_counts.append(i)
    test_set_data.append(set_time)
    test_get_data.append(get_time)
```
